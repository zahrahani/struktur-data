````python
# Implementasi Queue dengan menggunakan List

class QueueList:
    def __init__(self):
        self.queue =[]

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop(0)
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Contoh Penggunaan
queue = QueueList()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)

print("Queue setelah enqueue:", queue.queue)
print("Dequeued element     :", queue.dequeue())
print("Front element        :", queue.front())
print("Rear element         :", queue.rear())
print("Queue size           :", queue.size())
````

````python
from collections import deque

# Implementasi Queue menggunakan dequeu
class QueueDeque:
    def __init__(self):
        self.queue = deque()

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        return self.queue.popleft() if not self.is_empty() else None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Contoh Penggunaan
queue = QueueDeque()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)

print("Queue setelah enqueue:", list(queue.queue))
print("Dequeued element     :", queue.dequeue())
print("Front element        :", queue.front())
print("Rear element         :", queue.rear())
print("Queue size           :", queue.size())
````

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class QueueLinkedList:
    def __init__(self):
        self.front = self.rear = None

    def enqueue(self, data):
        new_node = Node(data)
        if self.rear is None:
            self.front = self.rear = new_node
            return
        self.rear.next = new_node
        self.rear = new_node

    def dequeue(self):
        if self.front is None:
            return None
        dequeued_element = self.front.data
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        return dequeued_element

    def is_empty(self):
        return self.front is None

    def front_element(self):
        return self.front.data if self.front else None

    def rear_element(self):
        return self.rear.data if self.rear else None

# Contoh penggunaan
queue = QueueLinkedList()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)

print("Dequeued element :", queue.dequeue())
print("Front element    :", queue.front_element())
print("Rear element     :", queue.rear_element())
````

```
SOAL LATIHAN 2: ANTREAN LAYANAN FOTOCOPY DI KAMPUS

Deskripsi:
Buatlah sistem antrean layanan fotocopy di kampus menggunakan Queue berbasis List,
di mana mahasiswa yang datang lebih awal akan mendapatkan layanan lebih dulu.

Ketentuan:
- 3 huruf pertama nama mahasiswa + 2 digit terakhir NPM = ID Pemesanan
- enqueue(order_id, num_pages) â†’ menambahkan pesanan fotocopy ke dalam antrean
  dengan jumlah halaman tertentu
- dequeue() â†’ melayani pesanan pertama dalam antrian
- front() â†’ menampilkan pesanan terakhir dalam antrean
- rear() â†’ menampilkan terakhir dalam antrean
- is_empty() â†’ mengecek apakah antrean kosong
- size() â†’ menghitung jumlah pesanan dalam antrean

Tambahan:
- Jika pesanan memiliki > 100 halaman, pesanan tersebut akan diproses dalam 2 tahap,
  sehingga tetap berada di antrean setelah tahap pertama selesai
- Fitur "Fast Track" â†’ pesanan < 10 halaman langsung diproses tanpa antre
```



````python
# Latihan 2 - Implementasi Queue berbasis List
# Sistem Layanan FotoCopy di Kampus

class AntreFC:
    def __init__(self, kapasitas = 100):
        self.queue = []
        self.kapasitas = kapasitas

    # Menambahkan pesanan fotocopy ke dalam antrean (enqueue())
    def tambahfc(self, order_id, num_pages):
        # Fitur Fast Track - tanpa antre
        if num_pages < 10:
            print(f"âœ… {order_id} â†’ {num_pages} halaman diproses tanpa antrean.\n")
            return   # Function Stop

        # Menambahkan ke antrean reguler
        if num_pages < self.kapasitas and num_pages >= 10:
            self.queue.append((order_id, num_pages))
            print(f"ğŸ“„ ID pemesanan {order_id} â†’ {num_pages} halaman")
            print("ğŸ“¥ Berhasil ditambahkan ke antrean.\n")
            return   # Function Stop

        # Menambah pesanan di antrean bertahap, 1 tahap = 100 halaman
        tahapke = 1
        while num_pages > self.kapasitas:
            self.queue.append((order_id, self.kapasitas))
            print(f"  ğŸ“¥ Tahap {tahapke}: {self.kapasitas} halaman berhasil masuk antrean.")
            num_pages -= self.kapasitas
            tahapke += 1
        if num_pages > 0:
            self.queue.append((order_id, num_pages))
            print(f"  ğŸ“¥ Tahap {tahapke}: {num_pages} halaman berhasil masuk antrean.\n")

    # Melayani/memproses pesanan pertama dalam antrean (dequeue())
    def layanifc(self):
        if not self.is_empty():
            pesanan = self.queue.pop(0)
            print(f"âœ… {pesanan[0]} â†’ {pesanan[1]} halaman selesai!\n")
        else:
            print("â€¼ï¸ Tidak ada yang bisa diproses â€¼ï¸\n")

    # Menampilkan pesanan pertama dalam antrean (front())
    def front(self):
        if not self.is_empty():
            print(f"ğŸ“„ Pesanan pertama: {self.queue[0][0]} â†’ {self.queue[0][1]} halaman\n")
        else:
            print("â€¼ï¸ Antrean kosong â€¼ï¸\n")

    # Menampilkan pesanan terakhir dalam antrean (rear())
    def rear(self):
        if not self.is_empty():
            print(f"ğŸ“„ Pesanan terakhir: {self.queue[-1][0]} â†’ {self.queue[-1][1]} halaman\n")
        else:
            print("â€¼ï¸ Antrean kosong â€¼ï¸\n")

    # Mengecek antrean kosong (is_empty())
    def is_empty(self):
        return len(self.queue) == 0

    # Menghitung jumlah pesanan dalam antrean (size())
    def banyakorder(self):
        print(f"ğŸ“‘ Jumlah pesanan dalam antrean: {len(self.queue)}\n")

# ======== Program Utama (Main) ======== #
# Membuat objek antre
antre = AntreFC(100)

# Menampilkan menu pilihan untuk user
print("       â­ Menu Pilihan â­")
print(" 1 -> tambah pesanan fotocopy")
print(" 2 -> layani/proses fotocopy")
print(" 3 -> lihat pesanan pertama")
print(" 4 -> lihat pesanan terakhir")
print(" 5 -> lihat banyak pesanan")
print(" 6 -> keluar dari program")
print()
while True:
    menu = input("Masukkan menu pilihan: ")

    # Memproses menu pilihan user
    if menu == '1':
        nama = input("â¡ï¸ Masukkan nama: ").upper()
        NPM = input("â¡ï¸ Masukkan NPM : ")
        order_id = nama[:3] + NPM[-2:]
        print(f"ğŸ“„ ID pemesanan : {order_id}")
        num_pages = int(input("ğŸ”¢ Masukkan jumlah halaman: "))
        antre.tambahfc(order_id, num_pages)
    elif menu == '2':
        antre.layanifc()
    elif menu == '3':
        antre.front()
    elif menu == '4':
        antre.rear()
    elif menu == '5':
        antre.banyakorder()
    elif menu == '6':
        print("ğŸ‘‹ Keluar dari program.")
        break
    else:
        print("âŒ Invalid input! Silakan pilih menu lagi.\n")
````

```
SOAL LATIHAN 3: ANTREAN LAYANAN PARKIR MALL

Deskripsi:
Buatlah sistem antrean kendaraan masuk ke parkiran mall menggunakan Queue berbasis List,
di mana kendaraan masuk dan keluar berdasarkan urutan kedatangannya.

Ketentuan:
- 2 huruf pertama kota lahir + 2 digit terakhir NPM = nomor plat kendaraan
- enqueue(vehicle_number) â†’ Menambahkan kendaraan ke dalam antrian parkir
- dequeue() â†’ mengeluarkan kendaraan yang lebih dulu datang dari parkiran
- front() â†’ menampilkan kendaraan pertama dalam antrian
- rear() â†’ menampilkan kendaraan terakhir dalam antrian.
- is_empty() â†’ mengecek apakah parkiran penuh
- size() â†’ menghitung jumlah kendaraan dalam antrian

Tambahan:
- Jika jumlah kendaraan melebihi kapasitas parkiran (ex: 15 kendaraan), kendaraan baru harus menunggu
  hingga ada kendaraan yang keluar
- Fitur "Parkir Prioritas" â†’ kendaraan ramah lingkungan (misalnya kendaraan listrik) mendapatkan
  prioritas masuk parkiran lebih dulu
```



pakai prinsip priority queue:

ğŸ”¹ Jika antrean masih ada, kendaraan pertama yang masuk akan keluar lebih dulu.

ğŸ”¹ Jika ada kendaraan listrik, mereka keluar lebih dulu karena dapat prioritas.

ğŸ”¹ Kalau antrean kosong, berarti tidak ada kendaraan yang bisa dikeluarkan.

````python
# Latihan 3
# Antrean Layanan Parkir Mall

class AntreParkir:
    def __init__(self, kapasitas = 15):
        self.queue = []
        self.kapasitas = kapasitas

    # Menambah kendaraan ke dalam antrean parkir (enqueue())
    def tambahkendaraan(self, platno):
        if len(self.queue) >= self.kapasitas:
            print("â›” Antrean penuh! Tunggu hingga ada space kosong â›”\n")
        else:
            self.queue.append(platno)
            print(f"ğŸ†— Kendaraan dengan plat {platno} berhasil ditambahkan ğŸ†—\n")

    # Mengeluarkan kendaraan yang lebih dulu datang ke parkiran (dequeue())
    def kendaraankeluar(self):
        if not self.isEmpty():
            antrean = self.queue.pop(0)
            print(f"â¬…ï¸ Kendaraan plat {antrean} keluar parkiran.\n")
        else:
            print("â€¼ï¸ Parkiran kosong â€¼ï¸\n")

    # Menampilkan kendaraan pertama dalam antrean (front())
    def front(self):
        if not self.isEmpty():
            print(f"ğŸ¯ Kendaraan antre pertama: plat {self.queue[0]} ğŸ¯\n")
        else:
            print("â€¼ï¸ Parkiran kosong â€¼ï¸\n")

    # Menampilkan kendaraan terakhir dalam antrean (rear())
    def rear(self):
        if not self.isEmpty():
            print(f"ğŸ¯ Kendaraan antre terakhir: plat {self.queue[-1]} ğŸ¯\n")
        else:
            print("â€¼ï¸ Parkiran kosong â€¼ï¸\n")

    # Mengecek parkiran kosong (isEmpty())
    def isEmpty(self):
        return len(self.queue) == 0

    # Menghitung jumlah kendaraan dalam antrean (size())
    def jumlahkendaraan(self):
        print(f"ğŸ…¿ï¸ Banyak kendaraan dalam antrean: {len(self.queue)} ğŸ…¿ï¸\n")

# ======== Program Utama (Main) ======== #
# Membuat objek antre
antre = AntreParkir(15)

# Menampilkan menu pilihan untuk user
print("      â­    Menu Pilihan    â­     ")
print(" 1 -> kendaraan masuk")
print(" 2 -> kendaraan keluar")
print(" 3 -> lihat kendaraan pertama dalam antrean")
print(" 4 -> lihat kendaraan terakhir dalam antrean")
print(" 5 -> jumlah kendaraan dalam antrean")
print(" 6 -> keluar dari program")
print()
while True:
    menu = input("Masukkan menu pilihan: ")

    # Memproses menu pilihan user
    if menu == '1':
        kota = input("â–«ï¸ Masukkan kota lahir: ").upper()
        NPM = input("â–«ï¸ Masukkan NPM       : ")
        platno = kota[:2] + NPM[-2:]
        jenis = input("â–«ï¸ Masukkan jenis kendaraan: ")

        if 'motor' in jenis:
            print(f"ğŸ›µ {jenis} plat {platno} ğŸ›µ")
        elif 'mobil' in  jenis:
            print(f"ğŸš— {jenis} plat {platno} ğŸš—")

        # Fitur Prioritas Parkir -> kendaraan ramah lingkungan
        if 'listrik' in jenis:
            print(f"ğŸ” Prioritas Parkir: Kendaraan {platno} masuk tanpa antre. ğŸ”\n")
            antre.queue.insert(0, platno)
        else:
            antre.tambahkendaraan(platno)
    elif menu == '2':
        antre.kendaraankeluar()
    elif menu == '3':
        antre.front()
    elif menu == '4':
        antre.rear()
    elif menu == '5':
        antre.jumlahkendaraan()
    elif menu == '6':
        print("ğŸ‘‹ Keluar dari program.")
        break
    else:
        print("âŒ Invalid input! Silakan pilih menu lagi.\n")
````

```
Buatlah sistem antrean pengambilan sertifikat di kampus menggunakan Queue berbasis List,
di mana mahasiswa yang telah menyelesaikan seminar atau pelatihan dapat mengambil sertifikat
berdasarkan urutan kedatangan.

Ketentuan:
- huruf pertama program studi + 2 digit terakhir NPM = nomor antrian mahasiswa
- enqueue(student_id) â†’ menambahkan mahasiswa ke dalam antrean pengambilan sertifikat
- dequeue() â†’ melayani mahasiswa yang lebih dulu datang
- front() â†’ menampilkan mahasiswa pertama dalam antrean
- rear() â†’ menampilkan mahasiswa terakhir dalam antrean
- is_empty() â†’ mengecek apakah antrean kosong
- size() â†’ menghitung jumlah mahasiswa dalam antrean

Tambahan:
- Jika jumlah mahasiswa dalam antrean > 20 orang, antrian akan dibagi menjadi 2 jalur,
  yaitu jalur mahasiswa reguler dan jalur mahasiswa berprestasi
- Mahasiswa yang pernah mengikuti kompetisi tingkat nasional bisa langsung mengambil
  sertifikat tanpa harus mengantri
```



````python
# Latihan 4
# Antrean Pengambilan Sertifikat di Kampus

class AntreSertif:
    def __init__(self, kapasitas = 20):
        self.queue = []     # antrean biasa
        self.reguler = []   # antrean mahasiswa reguler
        self.prestasi = []  # antrean mahasiswa berprestasi
        self.kapasitas = kapasitas

    # Menambahkan mahasiswa ke dalam antrean (enqueue())
    def tambahmhs(self, mahasiswa):
        if len(self.queue) < self.kapasitas:
            self.queue.append(mahasiswa)
            print(f"ğŸ‘¤ {mahasiswa} masuk ke antrean biasa ğŸ‘¤\n")
        else:
            prestasi = input("Mahasiswa berprestasi? (Y/N): ").upper()
            if prestasi == 'Y':
                self.prestasi.append(mahasiswa)
                print(f"ğŸ… {mahasiswa} masuk ke antrean jalur prestasi ğŸ…\n")
            elif prestasi == 'N':
                self.reguler.append(mahasiswa)
                print(f"ğŸ‘¥ {mahasiswa} masuk ke antrean jalur reguler ğŸ‘¥\n")

    # Melayani mahasiswa yang lebih dulu datang (dequeue())
    def layanimhs(self):
        if self.queue:
            mahasiswa = self.queue.pop(0)
            print(f"âœ… {mahasiswa} dari antrean biasa telah dilayani âœ…\n")
        elif self.prestasi:
            mahasiswa = self.prestasi.pop(0)
            print(f"ğŸ… {mahasiswa} dari jalur prestasi telah dilayani ğŸ…\n")
        elif self.reguler:
            mahasiswa = self.reguler.pop(0)
            print(f"â˜‘ï¸ {mahasiswa} dari jalur reguler telah dilayani â˜‘ï¸\n")
        else:
            print("â€¼ï¸ Antrean kosong â€¼ï¸\n")

    # Menampilkan mahasiswa pertama dalam antrean (front())
    def front(self):
        if self.queue:
            print(f"ğŸ‘¤ Antrean pertama: {self.queue[0]} (dari antrean biasa)\n")
        elif self.prestasi:
            print(f"ğŸ‘¤ Antrean pertama: {self.prestasi[0]} (dari jalur prestasi)\n")
        elif self.reguler:
            print(f"ğŸ‘¤ Antrean pertama: {self.reguler[0]} (dari jalur reguler)\n")
        else:
            print("â€¼ï¸ Antrean kosong â€¼ï¸\n")

    # Menampilkan pelanggan terakhir dalam antrean (rear())
    def rear(self):
        if self.queue:
            print(f"ğŸ‘¤ Antrean terakhir: {self.queue[-1]} (dari antrean biasa)\n")
        elif self.prestasi:
            print(f"ğŸ‘¤ Antrean terakhir: {self.prestasi[-1]} (dari jalur prestasi)\n")
        elif self.reguler:
            print(f"ğŸ‘¤ Antrean terakhir: {self.reguler[-1]} (dari jalur reguler)\n")
        else:
            print("â€¼ï¸ Antrean kosong â€¼ï¸\n")

    # Mengecek antrean kosong (is_empty())
    def is_empty(self):
        return len(self.queue) == 0 and len(self.reguler) == 0 and len(self.prestasi) == 0

    # Melihat jumlah mahasiswa dalam antrean (size())
    def jumlahmhsantre(self):
        total = len(self.queue) + len(self.reguler) + len(self.prestasi)
        print(f"ğŸ“ Mahasiswa dalam antrean {total}:")
        print(f"   - Biasa    : {len(self.queue)}")
        print(f"   - Reguler  : {len(self.reguler)}")
        print(f"   - Prestasi : {len(self.prestasi)}\n")

# ======== Program Utama (Main) ======== #
# Membuat objek antre
antre = AntreSertif()

# Menampilkan menu pilihan untuk user
print("      â­    Menu Pilihan    â­     ")
print(" 1 -> tambah antrean mahasiswa")
print(" 2 -> layani mahasiswa")
print(" 3 -> lihat antrean pertama")
print(" 4 -> lihat antrean terakhir")
print(" 5 -> jumlah mahasiswa dalam antrean")
print(" 6 -> keluar dari program")
print()
while True:
    menu = input("Masukkan menu pilihan: ")

    # Memproses menu pilihan user
    if menu == '1':
        prodi = input("â–«ï¸ Masukkan program studi: ").upper()
        NPM = input("â–«ï¸ Masukkan NPM          : ")
        nomor = prodi[:1] + NPM[-2:]

        # Layanan Prioritas -> ikut kompetisi tingkat nasional
        komnas = input("ğŸ† Mahasiswa pernah ikut kompetisi nasional? (Y/N): ").upper()
        if komnas == 'Y':
            print(f"ğŸ† Mahasiswa {nomor} ambil sertifikat tanpa antre. ğŸ†\n")
        else:
            antre.tambahmhs(nomor)
    elif menu == '2':
        antre.layanimhs()
    elif menu == '3':
        antre.front()
    elif menu == '4':
        antre.rear()
    elif menu == '5':
        antre.jumlahmhsantre()
    elif menu == '6':
        print("ğŸ‘‹ Keluar dari program.")
        break
    else:
        print("âŒ Invalid input! Silakan pilih menu lagi.\n")
````

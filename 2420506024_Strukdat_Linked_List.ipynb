```python
# Membuat Kelas Node dan Linked List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

## Implementasi Singly Linked List

```python
# Implementasi Singly Linked List (1)

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=' -> ')
            temp = temp.next
        print('None')
```

```python
# Implementasi Singly Linked List (2)

# Definisi Node dalam Linked List
class TaskNode:
    def __init__(self, task_name):
        self.task_name = task_name      # Menyimpan nama tugas
        self. next = None               # Pointer ke tugas berikutnya

# Definisi Linked List untuk Daftar Tugas
class TaskList:
    def __init__(self):
        self.head = None      # Awalnya tidak ada tugas (daftar kosong)

    # Menambahkan tugas ke Linked List
    def add_task(self, task_name):
        new_task = TaskNode(task_name)
        if not self.head:
            self.head = new_task    # Jika list kosong, tugas pertama menjadi head
        else:
            current = self.head
            while current.next:     # Mencari node terakhir dalam list
                current = current.next
            current.next = new_task   # Menambahkan tugas di akhir list

    # Menampilkan daftar tugas
    def show_tasks(self):
        if not self.head:
            print("\n Tidak ada tugas hari ini!")
            return

        print("\n Daftar Tugas Hari Ini:")
        current = self.head
        count = 1
        while current:
            print(f"{count}. {current.task_name}")
            current = current.next
            count += 1

# ------ Program Utama (Main) ------ #

# Membuat objek daftar tugas
task_list = TaskList()

# Meminta pengguna memasukkan jumlah tugas
num_tasks = int(input("Masukkan jumlah tugas hari ini: "))

# Loop untuk memasukkan tugas ke dalam Linked List
for i in range(1, num_tasks + 1):
    task_name = input(f"Masukkan tugas ke-{i}: ")
    task_list.add_task(task_name)

# Menampilkan hasil daftar tugas
task_list.show_tasks()
```

## Implementasi Doubly Linked List

```python
# Implementasi Doubly Linked List (1)

class DoublyNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = DoublyNode (data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end='  <-> ')
            temp = temp.next
        print('None')
```

```python
# Implementasi Doubly Linked List (2)

# Definisi Node dalam Doubly Linked List
class PageNode:
    def __init__(self, url):
        self.url = url    # URL halaman web
        self.next = None  # Pointer ke halaman berikutnya
        self.prev = None  # Pointer ke halaman sebelumnya

# Definisi Doubly Linked List untuk Riwayat Peramban
class BrowserHistory:
    def __init__(self):
        self.current = None   # Pointer ke halaman yang sedang dibuka

    # Menambahkan halaman baru ke dalam riwayat
    def visit_page(self, url):
        new_page = PageNode(url)
        if not self.current:
            self.current = new_page   # Jika riwayat kosong, halaman ini menjadi halaman pertama
        else:
            new_page.prev = self.current
            self.current.next = new_page
            self.current = new_page   # Pindahkan posisi ke halaman baru

    # Menampilkan riwayat kunjungan
    def show_history(self):
        if not self.current:
            print("\n Riwayat kosong.")
            return

        print("\n Riwayat Peramban Web:")
        temp = self.current
        while temp.prev:    # Kembali ke halaman pertama
            temp = temp.prev

        count = 1
        while temp:
            print(f"{count}. {temp.url}")
            temp = temp.next
            count += 1

    # Navigasi mundur ke halaman sebelumnya
    def go_back(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f"\n Anda kembali ke: {self.current.url}")
        else:
            print("\n Tidak bisa kembali, ini adalah halaman pertama!")

    # Navigasi maju ke halaman berikutnya
    def go_forward(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"\n Anda maju ke: {self.current.url}")
        else:
            print("\n Tidak bisa maju, ini adalah halaman terakhir!")

# ------ Program Utama (Main) ------ #

# Membuat objek riwayat peramban
history = BrowserHistory()

# Meminta pengguna memasukkan jumlah halaman yang dikunjungi
num_pages = int(input("\nMasukkan jumlah halaman yang dikunjungi: "))

# Loop untuk memasukkan URL halaman ke dalam Doubly Linked List
for i in range(1, num_pages + 1):
    url = input(f"Masukkan URL halaman ke-{i}: ")
    history.visit_page(url)

# Menampilkan riwayat yang telah dimasukkan
history.show_history()

# Menu riwayat yang telah dimasukkan
while True:
    command = input("\n Tekan 'b' untuk kembali, Tekan 'f' untuk maju, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()

    if command == 'b':
        history.go_back()
    elif command == 'f':
        history.go_forward()
    elif command == 'x':
        print("\n Keluar dari riwayat peramban.")
        break
    else:
        print("\n Perintah tidak dikenal.")
```

```python
# Latihan 2
# Buatlah program yang menampilkan riwayat navigasi folder dalam File Explorer
# menggunakan Doubly Linked List. Jumlah folder yang Anda navigasikan sesuai
# dengan digit terakhir NPM Anda!

# NPM = 2420506024 >> Jadi, jumlah folder yang dinavigasikan ada 4

# Definisi Node dalam Doubly Linked List
class FolderNode:
    def __init__(self, nama_folder):
        self.nama_folder = nama_folder
        self.next = None
        self.prev = None

# Definisi Doubly Linked List untuk Riwayat Folder
class HistoryFile:
    def __init__(self):
        self.current = None

    # Menambahkan folder baru ke daftar riwayat folder
    def visit_folder(self, nama_folder):
        new_folder = FolderNode(nama_folder)
        if not self.current:
            self.current = new_folder   # Jika riwayat kosong, folder ini menjadi node pertama
        else:
            new_folder.prev = self.current
            self.current.next = new_folder
            self.current = new_folder   # Pindahkan posisi ke folder baru
        print(f"\n Anda berada di folder: {self.current.nama_folder}")

    # Menampilkan riwayat navigasi folder
    def show_history(self):
        if not self.current:
            print("\nRiwayat kosong.")
            return

        print("\nRiwayat folder di File Explorer :")
        temp = self.current
        while temp.prev:          # Kembali ke folder pertama
            temp = temp.prev

        count = 1
        while temp:               # Membuat daftar riwayat navigasi folder
            print(f"{count}. {temp.nama_folder}")
            temp = temp.next
            count += 1

    # Navigasi mundur ke folder sebelumnya
    def go_back(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f"\n Anda kembali ke: {self.current.nama_folder}")
        else:
            print("\n Tidak bisa kembali, ini adalah folder pertama!")

    # Navigasi maju ke folder berikutnya
    def go_forward(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"\n Anda maju ke: {self.current.nama_folder}")
        else:
            print("\n Tidak bisa maju, ini adalah folder terakhir!")


# ------ Program Utama (Main) ------ #

# Membuat objek riwayat navigasi folder
history = HistoryFile()

# Meminta pengguna memasukkan NPM
npm = input("Masukkan NPM : ")

# Menetapkan digit terakhir NPM = jumlah folder yang dibuka
jumlahfolder = int(npm[-1])
print(f"\nJumlah folder yang dinavigasikan: {jumlahfolder}")

# Loop untuk memasukkan folder ke dalam Doubly Linked List
for i in range(1, jumlahfolder + 1):
    folder = input(f"Masukkan folder ke-{i}: ")
    history.visit_folder(folder)

# Menampilkan riwayat navigasi folder
history.show_history()

# Menu riwayat yang telah dimasukkan
while True:
    command = input("\n Tekan 'b' untuk kembali, Tekan 'f' untuk maju, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()

    if command == 'b':
        history.go_back()
    elif command == 'f':
        history.go_forward()
    elif command == 'x':
        print("\n Keluar dari File Explorer.")
        break
    else:
        print("\n Perintah tidak dikenal.")
```

## Implementasi Circular Linked List

```python
# Implementasi Circular Linked List (1)

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(sel, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = new_node
        new_node.next = self.head

    def display(self):
        if not self.head:
            return
        temp = self.head
        while True:
            print(temp.data, end=' -> ')
            temp = temp.next
            if temp == self.head:
                break
        print('(Back to Head)')
```

```python
# Implementasi Circular Linked List (2)

# Definisi Node dalam Circular Linked List
class SongNode:
    def __init__(self, title):
        self.title = title   # Menyimpan judul lagu
        self.next = None     # Pointer ke lagu berikutnya

# Definisi Circular Linked List untuk Playlist Lagu
class Playlist:
    def __init__(self):
        self.current = None       # Lagu yang sedang diputar

    # Menambahkan lagu ke dalam playlist
    def add_song(self, title):
        new_song = SongNode(title)
        if not self.current:
            self.current = new_song
            self.current.next = self.current    # Menghubungkan ke dirinya sendiri (Circular)
        else:
            temp = self.current
            while temp.next != self.current:     # Mencari lagu terakhir
                temp = temp.next
            temp.next = new_song    # Menambahkan lagu di akhir
            new_song.next = self.current    # Menghubungkan kembali ke awal

    # Menampilkan daftar lagu dalam playlist
    def show_playlist(self):
        if not self.current:
            print("\n Playlist kosong!")
            return

        print("\n Playlist Lagu Anda:")
        temp = self.current
        count = 1
        while True:
            print(f"{count}. {temp.title}")
            temp = temp.next
            count += 1
            if temp == self.current:    # Berhenti jika sudah kembali ke awal
                break

    # Memutar Lagu Berikutnya dalam Playlist
    def next_song(self):
        if self.current:
            self.current = self.current.next
            print(f"\n Memutar: {self.current.title}")

# ====== Program Utama (Main) ====== #

# Membuat Objek Playlist
playlist = Playlist()

# Meminta Pengguna Memasukkan Jumlah Lagu
num_songs = int(input("\n Masukkan jumlah lagu dalam playlist: "))

# Loop untuk Memasukkan Lagu ke Dalam Circular Linked List
for i in range(1, num_songs + 1):
    title = input(f"Masukkan judul lagu ke-{i}: ")
    playlist.add_song(title)

# Menampilkan Daftar Lagu
playlist.show_playlist()

# Menu Navigasi Pengguna
while True:
    command = input("\n Tekan 'n' untuk lagu berikutnya, Tekan 'x' untuk keluar.\n Masukkan perintah: ").lower()

    if command == 'n':
        playlist.next_song()
    elif command == 'x':
        print("\n Keluar dari pemutar musik.")
        break
    else:
        print("\n PErintah tidak dikenal.")
```

## Operasi Dasar pada Linked List

```python
# Definisi Node dalam Singly Linked List
class Node:
    def __init__(self, data):
        self.data = data  # Menyimpan nilai node
        self.next = None  # Pointer ke node berikutnya

# Definisi Singly Linked List dengan Berbagai Operasi
class LinkedList:
    def __init__(self):
        self.head = None  # Awalnya, daftar kosong

    # Menambahkan elemen di awal
    def add_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Menambahkan elemen di akhir
    def add_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Menambahkan elemen di posisi tertentu
    def add_at_position(self, position, data):
        if position < 0:
            print(" Posisi tidak valid!")
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        temp = self.head
        for _ in range(position - 1):
          if not temp:
              print(" Posisi melebihi panjang daftar!")
              return
          temp = temp.next
        new_node.next = temp.next
        temp.next = new_node

    # Menghapus elemen di awal
    def delete_at_beginning(self):
        if not self.head:
            print(" Daftar kosong, tidak ada yang bisa dihapus!")
            return
        self.head = self.head.next

    # Menghapus elemen di akhir
    def delete_at_end(self):
        if not self.head:
            print(" Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None

    # Menghapus elemen di posisi tertentu
    def delete_at_position(self, position):
        if not self.head:
            print(" Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if position == 0:
             self. head = self.head.next
             return
        temp = self.head
        for _ in range(position - 1):
            if not temp or not temp.next:
                print(" Posisi tidak ditemukan!")
                return
            temp = temp.next
        temp.next = temp.next.next

    # Mencari elemen dalam Linked List
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                print(f" Elemen {key} ditemukan dalam daftar!")
                return
            temp = temp.next
        print(f" Elemen {key} tidak ditemukan dalam daftar.")

    # Membalik (reverse) Linked List
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    # Menampilkan Linked List
    def display(self):
        if not self.head:
            print(" Daftar kosong!")
            return
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

# ====== Program Utama (main) ====== #

# Membuat objek Linked List
ll = LinkedList()

# Meminta pengguna memasukkan elemen awal
num_elements = int(input("\nMasukkan jumlah awal elemen dalam daftar: "))
for i in range(num_elements):
    data = int(input(f"Masukkan data ke{i+1}: "))
    ll.add_at_end(data)

# Menu operasi Linked List
while True:
    print("\n Pilih operasi:")
    print("1. Tambah elemen di awal")
    print("2. Tambah elemen di akhir")
    print("3. Tambah elemen di posisi tertentu")
    print("4. Hapus elemen di awal")
    print("5. Hapus elemen di akhir")
    print("6. Hapus elemen di posisi tertentu")
    print("7. Cari elemen dalam daftar")
    print("8. Balikkan daftar")
    print("9. Tampilkan daftar")
    print("0. Keluar")

    choice = int(input("\nMasukkan pilihan: "))

    if choice == 1:
        data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
        ll.add_at_beginning(data)
    elif choice == 2:
        data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
        ll.add_at_end(data)
    elif choice == 3:
        position = int(input("Masukkan posisi untuk menambahkan data:"))
        data = int(input("Masukkan data yang ingin ditambahkan: "))
        ll.add_at_position(position, data)
    elif choice == 4:
        ll.delete_at_beginning()
    elif choice == 5:
        ll.delete_at_end()
    elif choice == 6:
        position = int(input("Masukkan posisi elemen yang ingin dihapus:"))
        ll.delete_at_position(position)
    elif choice == 7:
        key = int(input("Masukkan elemen yang ingin dicari: "))
        ll.search(key)
    elif choice == 8:
        ll.reverse()
        print(" Daftar berhasil dibalik!")
    elif choice == 9:
        ll.display()
    elif choice == 0:
        print("\n Keluar dari program")
        break
    else:
        print("\n Pilihan tidak valid, silakan coba lagi!")
```
